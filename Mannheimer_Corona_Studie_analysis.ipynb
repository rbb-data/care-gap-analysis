```python
#we create one dataframe to just check out before and after pandemic childcare
basedf=pd.DataFrame()
#so we iterate and check for each form of childcare (a to e) whether the item is checked
basedf["childcare_vorher"]="0"
for buchstabe in ["a","b","c","d","e"]:
    print("SCFX002_"+str(buchstabe))
    currentdf=pd.DataFrame()
    #and write out id and week
    currentdf["id"]=data[data["SCFX002_"+buchstabe]=='1. item checked']["id_g"].astype(str).copy()
    currentdf["woche"]=data[data["SCFX002_"+buchstabe]=='1. item checked']["Woche"].astype(str).copy()
    #plus the letter of this type of childcare
    currentdf["childcare_vorher"]=str(buchstabe)
    #and concat
    basedf=pd.concat([basedf,currentdf])
#we drop duplicates
basedf=basedf.drop_duplicates()

#then we do the same for types of childcare since pandemic 
basedf_two=pd.DataFrame()
basedf_two["childcare_nachher"]="0"
for buchstabe in ["a","b","c","d","e","f","g","h"]:
    print("SCFX003_"+str(buchstabe))
    currentdf=pd.DataFrame()
    currentdf["id"]=data[data["SCFX003_"+buchstabe]=='1. item checked']["id_g"].astype(str).copy()
    currentdf["woche"]=data[data["SCFX003_"+buchstabe]=='1. item checked']["Woche"].astype(str).copy()

    currentdf["childcare_nachher"]=str(buchstabe)
    basedf_two=pd.concat([basedf_two,currentdf])
basedf_two=basedf_two.drop_duplicates()
basedf_two
#then we merge this by using id and week
childcare_df=pd.merge(basedf,basedf_two,on=["id","woche"])
childcare_df.id=childcare_df.id.astype(float)
childcare_df.woche=childcare_df.woche.astype(float)
childcare_df
```

```python
#also we create two lists of ids based on whether someone has a partner or not
singlelist=extract[(extract["partner"]=="0")]["id_g"].unique().tolist()
partnerlist=extract[(extract["partner"]=="1")]["id_g"].unique().tolist()
```

```python
#then we create a dictionary of datasubsets based on whether parents are single or partnered and whether they are male or female
df_dict={}
df_dict["single_mothers"]=extract[(extract["gender_19"]=="2. weiblich")&(extract["Kinder_nachher"]>0)&(extract["id_g"].isin(singlelist))]
df_dict["single_fathers"]=extract[(extract["gender_19"]=="1. mÃ¤nnlich")&(extract["Kinder_nachher"]>0)&(extract["id_g"].isin(singlelist))]
df_dict["partner_mothers"]=extract[(extract["gender_19"]=="2. weiblich")&(extract["Kinder_nachher"]>0)&(extract["id_g"].isin(partnerlist))]
df_dict["partner_fathers"]=extract[(extract["gender_19"]=="1. mÃ¤nnlich")&(extract["Kinder_nachher"]>0)&(extract["id_g"].isin(partnerlist))]
```

```python
print(len(df_dict["single_mothers"]["id_g"].unique()))
len(df_dict["single_fathers"]["id_g"].unique())
```

```python
print(len(df_dict["partner_mothers"]["id_g"].unique()))
print(len(df_dict["partner_fathers"]["id_g"].unique()))
```

```python
#we build a list of metrics that we have values for before the pandemic and that are numeric
print(df_dict["single_mothers"].columns[df_dict["single_mothers"].columns.str.contains("_vorher")])
metriclist=df_dict["single_mothers"].columns[df_dict["single_mothers"].columns.str.contains("_vorher")][1:6]

metriclist
```

```python
#now we want a dataframe where we can trace back how the values have changed from before the pandemic
concat_metrics_subset=pd.DataFrame()

#we iterate through these metrics
for vorher in metriclist:
    nachher=(vorher.replace("_vorher","_nachher"))
    print(nachher)
    concat_subset=pd.DataFrame()
    #we iterate through the different subsets
    for key in df_dict.keys():
        #we write into a dictionary that will store the mean values
        dict_of_values={}
        #and we iterate through the weeks from 1 to 16
        for i in range(16):
            #if we have for this metric and for this subset not a nan for mean
            if math.isnan(df_dict[key][df_dict[key]["Woche"]==i][nachher].mean())==False:
                #we write this to our dictionary of values
                dict_of_values[i]=(df_dict[key][df_dict[key]["Woche"]==i][nachher].mean())
        #we also write the mean value for before the pandemic
        dict_of_values["vorher"]=(pd.to_numeric(df_dict[key][vorher]).mean())
        #and then we create a dataframe from this dictionary
        df=pd.DataFrame(dict_of_values, index=[0])
        #and write into a column which subset this is
        df["subset"]=key
        #and then concat this df to other dfs of other subsets
        concat_subset=pd.concat([concat_subset,df])
    #and once we are done through with all subsets
    concat_subset["metric"]=nachher.replace("_nachher","")
    #we add this df for one metric and all subsets to the greater df
    concat_metrics_subset=pd.concat([concat_metrics_subset,concat_subset])
concat_metrics_subset
```
