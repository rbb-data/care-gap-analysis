```python
#loading libraries
import glob
import pandas as pd
```

```python
#these are all the files that are weighted and differentiate between mothers or fathers, partnered or single
filenames=glob.glob("../RawData/soep_cov_weighted/*")
filenames
```

```python
#we get the name of the variable from the filename
metric=filenames[0][38:-26]
#we load in the very first file
data=pd.read_csv(filenames[0])

#we retrieve the mean and the se by parental status and for each year
meandf=data.pivot(index='syear', columns='parental_status', values='mean').reset_index()
sedf=data.pivot(index='syear', columns='parental_status', values='se').reset_index()

#then we merge these variables side by side
mergemeanse=pd.merge(meandf,sedf,on="syear",suffixes=["_mean","_se"])

#then we write the metric into one of the columns
mergemeanse["cat"]=metric


#and then bring this into a form by resorting the columns
basedf=mergemeanse[["syear","cat","mother with partner_mean","mother with partner_se","father with partner_mean","father with partner_se","single mom_mean","single mom_se"]]
print(basedf.head())
```

```python
#then we iterate throught all the files we have except the first one
for file in filenames[1:]:
    metric=file[38:-26]
    data=pd.read_csv(file)
    meandf=data.pivot(index='syear', columns='parental_status', values='mean').reset_index()
    sedf=data.pivot(index='syear', columns='parental_status', values='se').reset_index()

    mergemeanse=pd.merge(meandf,sedf,on="syear",suffixes=["_mean","_se"])

    mergemeanse["cat"]=metric

    mergemeanse=mergemeanse[["syear","cat","mother with partner_mean","mother with partner_se","father with partner_mean","father with partner_se","single mom_mean","single mom_se"]]
    print(mergemeanse.head())
    basedf=pd.concat([basedf,mergemeanse],axis=0)
mergemeanse
```

```python
#we sum up for each columns the sum of things that we considers as care work
careworklist=["care","errands","repair","chores","kids"]
sumcare=basedf[basedf.cat.isin(careworklist)].groupby("syear").sum().reset_index()
sumcare["cat"]="sum_of_carework"
sumcare=sumcare[basedf.columns]
alldf=pd.concat([basedf,sumcare])
```

```python
#we do the same for work

workworklist=["job","training"]
sumwork=basedf[basedf.cat.isin(workworklist)].groupby("syear").sum().reset_index()
sumwork["cat"]="sum_of_workwork"
sumwork=sumwork[basedf.columns]
alldf_2=pd.concat([alldf,sumwork])
alldf_2
```

```python
#we do the same for freetime

freetimelist=["hobbies","sports"]
sumfree=basedf[basedf.cat.isin(freetimelist)].groupby("syear").sum().reset_index()
sumfree["cat"]="sum_of_freetime"
sumfree=sumfree[basedf.columns]
alldf_3=pd.concat([alldf_2,sumfree])
```

```python
#we retrieve the data for 2019 and for 2020
the_2019_data=alldf_3[alldf_3["syear"].astype(str).isin(["2019"])][["cat","mother with partner_mean","father with partner_mean","single mom_mean"]]
the_2020_data=alldf_3[alldf_3["syear"].astype(str).isin(["2020"])][["cat","mother with partner_mean","father with partner_mean","single mom_mean"]]
```

```python
#and the merge the data for 2019 and 2020 and again resort the data
resorted_data=pd.merge(the_2019_data,the_2020_data,on="cat",suffixes=["_2019","_2020"])
columns_list=['cat',
 'mother with partner_mean_2019',
 'mother with partner_mean_2020',
 'father with partner_mean_2019',
 'father with partner_mean_2020',
 'single mom_mean_2019',
 'single mom_mean_2020']
resorted_data[columns_list].to_csv("soep_cov_2020_weighted_but_not_panel.csv")
```

```python
diffdata=pd.DataFrame()
diffdata["cat"]=resorted_data["cat"]
diffdata["mother with partner_mean_diff"]=resorted_data["mother with partner_mean_2020"]-resorted_data["mother with partner_mean_2019"]
diffdata["father with partner_mean_diff"]=resorted_data["father with partner_mean_2020"]-resorted_data["father with partner_mean_2019"]
diffdata["single mom_mean_diff"]=resorted_data["single mom_mean_2020"]-resorted_data["single mom_mean_2019"]
diffdata
```
