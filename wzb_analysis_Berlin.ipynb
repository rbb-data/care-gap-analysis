```python
#we define a list of metrics
criterialist=[]
for i in extract.columns[10:].tolist():
    criterialist.append(i.replace("_nachher","").replace("_vorher",""))
criterialist=set(criterialist)
criterialist
```

```python
#we build a dataframe
alldf=pd.DataFrame()
#metric by metric
for criterion in criterialist:
    print(criterion)
    #and for each subset
    for subset in [single_mothers_wave3,partner_fathers_wave3,partner_mothers_wave3]:
        #get the gender, partner status, length of df
        gender=subset["Gender"].unique()[0]
        partner=subset["partner"].unique()[0]
        wave=subset["wave"].unique()[0]
        subsetsize=len(subset)

        #substract the value counts since the pandemic if this metric 
        #(how often a certain answer was given in percent)
        #from the value counts before the pandemic
        series=(subset[criterion+"_nachher"].value_counts(normalize=True)-subset[criterion+"_vorher"].value_counts(normalize=True))
        
        #save the indices and values of this analysis as lists
        indices=(series.index.tolist())
        values=(series.values.tolist())
        
        #and now to have a value or string for the other columns
        #that are added to this analysis, we do one-liner-for-loops
        metricslist=[criterion for i in range(len(series.values.tolist()))]
        genderlist=[gender for i in range(len(series.values.tolist()))]
        partnerlist=[partner for i in range(len(series.values.tolist()))]
        sizelist=[subsetsize for i in range(len(series.values.tolist()))]
        wavelist=[wave for i in range(len(series.values.tolist()))]
        
        #and then we build a dataframe from these lists
        newdf=pd.DataFrame({"indices":indices,"values":values,"metric":metricslist,"subsetsize":sizelist,"gender":genderlist,"partner":partnerlist,"wave":wavelist})
        alldf=pd.concat([alldf,newdf])
```
